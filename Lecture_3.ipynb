#### Gaussian Linear Regression

1. yi ~ Norm(mui,sigma)
2. mui = alpha + beta*(xi-mux)
3. alpha ~ Norm()
4. beta ~ Norm()
5. sigma ~ Unif(0,50)

```python
x = np.arange(0,50)
```

```python
y = sc.distributions.norm(162+3*(x-x.mean()),2)
```

```python
tmp = []
for i in range(100):
    tmp.append([x,y.rvs(50)])
for e in tmp:
    sns.scatterplot(e[0],e[1])
```

```python
beta = sc.distributions.norm(loc=1,scale=10)

alpa = sc.distributions.norm(loc=150,scale=30)
```

```python
final = []
for i in range(100):
    mui = alpa.rvs(1) + beta.rvs(1)*(x-x.mean())
    final.append([x,mui])

for e in final:
    sns.scatterplot(e[0],e[1])
```
